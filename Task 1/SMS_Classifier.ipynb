# Text Preprocessing

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
corpus=[] #It is text data

for i in range(df.shape[0]):
    features=re.sub('[^a-zA-Z]',' ',df['msg'][i])
    features=features.lower()
    corpus.append(features)
```

```python
df['features']=corpus
df.head()
```

# Defining Features and Labels

```python
x=df['features']
y=df['is_spam']
```

```python
cvector=CountVectorizer()
cvdf=cvector.fit_transform(x)
```

```python
cvdf.toarray()
```

```python
sm=SMOTE()
x_sm,y_sm=sm.fit_resample(cvdf,y)
```

```python
y_sm.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_sm,y_sm,test_size=0.2,random_state=0)
```

```python
svm=SVC(kernel='linear')
```

```python
svm.fit(x_train,y_train)
```

```python
#wait
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
def check_spam(input_text):
    yp = svm.predict(cvector.transform([input_text]))
    return  yp
```

```python
#Testing

test_msg = input("Enter the mail: ")
is_spam = check_spam(test_msg)
if is_spam == 1:
  print("It's a SPAM message, rest in peace.")
else:
  print("No worries, It's not a SPAM mail")
```
